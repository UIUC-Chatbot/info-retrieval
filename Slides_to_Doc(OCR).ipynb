```python
# import the necessary packages
from PIL import Image
import pytesseract
import argparse
import cv2
import os
```

OCR

```python
## Converts image to PIL image and then uses the PIL image for OCR and creates a new doc
def image_pil_doc(img_path,image_folder,slide,pil_folder,doc_folder, pre_process="thresh"): 
  # load the example image and convert it to grayscale
  image = cv2.imread(img_path)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  # check to see if we should apply thresholding to preprocess the
  # image
  if pre_process == "thresh":
    gray = cv2.threshold(gray, 0, 255,
      cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
  # make a check to see if median blurring should be done to remove
  # noise
  elif pre_process == "blur":
    gray = cv2.medianBlur(gray, 3)
  # write the grayscale image to disk as a temporary file so we can
  # apply OCR to it
  filename = os.path.join(pil_folder, image_folder + '_' +slide+'.png')
  print(filename)
  cv2.imwrite(os.path.join(pil_folder, image_folder + '_' +slide+'.png'), gray)

  # load the image as a PIL/Pillow image, apply OCR, and then delete
  # the temporary file
  text = pytesseract.image_to_string(Image.open(filename))
  os.remove(filename)
  # print(text)
  # print(type(text))

  ## Creating a text document
  doc_path = os.path.join(doc_folder, image_folder + '_' +slide+'.txt')
  text_file = open(doc_path, "w")
  text_file.write(text)
  text_file.close()
```

```python
path_of_ppt_folders = '/content/lecture_slides'
path_of_pil_images = '/content/PIL_images'
path_of_doc_files = '/content/documents'
ppts = list(os.listdir(path_of_ppt_folders))

for i in ppts:
            # Get the path of the folder containing the ppt images
            folder_name = i
            images = list(os.listdir(os.path.join(path_of_ppt_folders, folder_name)))  
            
            for img in images:
              img_path = os.path.join(path_of_ppt_folders, folder_name,img)
              slide = img.split('.')[0]
              image_pil_doc(img_path,folder_name,slide,path_of_pil_images,path_of_doc_files, pre_process="thresh")
```
